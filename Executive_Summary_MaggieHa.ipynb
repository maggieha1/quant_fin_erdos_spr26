## Executive Summary: Pricing and Delta-Hedging American Options

This project established a quantitative model for pricing and delta-hedging American options under the standard Black-Scholes assumption of Geometric Brownian Motion (GBM) for underlying stock prices. The core objective was to backtest this theoretical framework against historical market data to quantify model risk and understand discrepancies arising from the breakdown of assumptions like constant volatility and continuous trading in real-world scenarios.

**Key Components and Methodology:**

1.  **Pricing Model:** Due to the absence of a closed-form solution for American options, the Cox-Ross-Rubinstein Binomial Tree model was employed. This model discretizes the GBM process to determine option values through backward induction, accounting for early exercise possibilities.

2.  **Volatility Modeling:** Both historical (realized) and implied (market-derived) volatilities were discussed. The project implemented methods to calculate historical volatility from past stock returns and used a Newton-Raphson numerical method to extract implied volatility from market option prices.

3.  **Hedging Strategy:** A delta-neutral hedging strategy was outlined, involving the calculation of an option's Delta (sensitivity to stock price changes) to dynamically rebalance a portfolio. The aim was to offset directional risk by trading the underlying stock.

4.  **Empirical Evaluation & Backtesting:** A delta-hedging simulation was performed using historical stock prices and simulated market option prices, crucially introducing a volatility surface shift midway to mimic real-world market dynamics. The hedge was maintained using a constant volatility assumption, while market option prices reflected a sudden change in implied volatility.

**Results:**

The backtest revealed a significant hedging error or bleed of approximately **$-0.21** over the simulation period. This error, quantified by the final portfolio profit and loss, demonstrates the practical challenges and costs associated with maintaining a delta-neutral hedge when real-world market conditions (like sudden volatility shifts and discrete rebalancing) deviate from the idealized Black-Scholes assumptions. The simulation effectively illustrated that even with a theoretically sound model, real-world frictions lead to tracking errors.

**Conclusion:**

The project shows that while GBM and binomial tree models are fundamental tools in financial mathematics, effective risk management in live markets necessitates empirical calibration and models that can adapt to the complex, dynamic nature of real world volatility, such as those accounting for stochastic or local volatility. The observed hedging error highlights the importance of understanding and quantifying model risk beyond theoretical perfection.
